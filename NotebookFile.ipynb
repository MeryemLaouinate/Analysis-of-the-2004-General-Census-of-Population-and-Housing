```python
#12.Création de la fonction « codeProvince » qui permet d’extraire le code de la province à partir du code de la commune. 

def codeProvince(s):
    cp = s.split('.')[0]
    return cp
#Ajout de la colonne « Code_Province »
dfc['Code_Province']=dfc['Code_Commune'].apply(codeProvince)
dfc
```

```python
#13.Création de la fonction « nomProvince » qui permet de déterminer le nom de la province à partir de son code présent dans le fichier csv.
df = pd.read_csv('Data_Projet_Python/Code_Nom_Province.csv')


dataf = pd.DataFrame(df, columns = ['Code_Province' , 'Nom_Province'])

def nomProvince(Code_Province):  
   datap = dataf.loc[dataf['Code_Province'] == int(Code_Province)]
   try:
      nom_Province = datap['Nom_Province'].values[0]
   except:
      nom_Province = "Inexistant"  
   return nom_Province
 
#Ajout de la colonne « Nom_Province » aux données « dfc » 

dfc['Nom_Province']= dfc['Code_Province'].apply(nomProvince)
dfc

```

```python
#14.1 Mettre les deux colonnes « Code_Province » et « Code_Commune » comme index pour « dfc »
dfc.set_index(['Code_Province','Code_Commune'],inplace=True)
dfc
```
